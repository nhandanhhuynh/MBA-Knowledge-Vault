# Biểu đồ minh họa – Tình huống Xe điện Alpha

Notebook tạo 5 biểu đồ kinh tế lý thuyết cho bài trình bày phân tích tình huống xe điện tại quốc gia Alpha.

- `01_supply_demand_shift.png` – Cung-Cầu dịch chuyển (Câu 1)
- `02_substitute_complement.png` – Hàng thay thế vs Hàng bổ sung (Câu 2)
- `03_mc_decision.png` – MR=MC quyết định sản lượng (Câu 3)
- `04_differentiation.png` – Khác biệt hóa và độ co giãn (Câu 4)
- `05_market_structure.png` – Chuyển đổi cấu trúc thị trường (Câu 4)

```python
import matplotlib.pyplot as plt
import matplotlib.patches as mpatches
import numpy as np
import os

IMG_DIR = "images"
os.makedirs(IMG_DIR, exist_ok=True)

plt.rcParams['font.size'] = 11
```

## 1. Cung-Cầu dịch chuyển (Câu 1)

Cầu dịch mạnh sang phải (thu nhập ↑, trợ cấp, thuế xe xăng). Cung bị kìm (Lithium ↑ vs Công nghệ pin ↑). Kết quả: giá cân bằng TĂNG.

```python
fig, ax = plt.subplots(figsize=(7.5, 5))
q = np.linspace(0, 10, 100)

ax.plot(q, 10-q, 'b--', lw=1.8, label='D₁')
ax.plot(q, 12-q, 'b-', lw=2.5, label='D₂ (Cầu tăng)')
ax.plot(q, 2+0.8*q, 'r--', lw=1.8, label='S₁')
ax.plot(q, 3.5+0.8*q, 'r-', lw=2.5, label='S₂ (Cung bị kìm)')

qe1 = (10-2)/1.8; pe1 = 10-qe1
qe2 = (12-3.5)/1.8; pe2 = 12-qe2
ax.plot(qe1, pe1, 'ko', ms=9)
ax.plot(qe2, pe2, 'go', ms=11)

ax.annotate('E₁', (qe1, pe1), xytext=(-15, 10), textcoords='offset points', fontsize=12, fontweight='bold')
ax.annotate('E₂ (giá cao hơn)', (qe2, pe2), xytext=(8, 8), textcoords='offset points', fontsize=11, fontweight='bold', color='green')

ax.axhline(pe1, color='gray', ls=':', alpha=.4)
ax.axhline(pe2, color='green', ls=':', alpha=.4)

ax.set_xlabel('Sản lượng (Q)', fontsize=12)
ax.set_ylabel('Giá (P)', fontsize=12)
ax.set_title('Kết luận Câu 1: Cầu mạnh + Cung yếu → Giá tăng', fontsize=12, fontweight='bold')
ax.legend(fontsize=10, loc='upper right')
ax.set_xlim(0, 10); ax.set_ylim(0, 12)
ax.grid(True, alpha=.25)
plt.tight_layout()
plt.savefig(os.path.join(IMG_DIR, '01_supply_demand_shift.png'), dpi=150, bbox_inches='tight')
plt.show()
```

## 2. Hàng thay thế vs Hàng bổ sung (Câu 2)

- Trái: Thuế xe xăng (hàng thay thế đắt) → Cầu EV tăng
- Phải: Điện sạc đắt (hàng bổ sung) → Cầu EV giảm, lợi ích cận biên bị xói mòn

```python
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(11, 4.5))
q = np.linspace(0, 10, 100)

# Hàng thay thế
ax1.plot(q, 8-0.8*q, 'b--', lw=1.8, label='D₁')
ax1.plot(q, 10-0.8*q, 'b-', lw=2.5, label='D₂')
ax1.plot(q, 1+0.7*q, 'r-', lw=2, label='S')
ax1.annotate('', xy=(7, 5), xytext=(5, 5), arrowprops=dict(arrowstyle='->', color='blue', lw=2.5))
ax1.set_title('Xe xăng đắt (thuế) → Cầu EV TĂNG\n[Hàng thay thế]', fontsize=11, fontweight='bold')
ax1.set_xlabel('Q'); ax1.set_ylabel('P')
ax1.legend(fontsize=9)
ax1.set_xlim(0, 10); ax1.set_ylim(0, 11)
ax1.grid(True, alpha=.25)

# Hàng bổ sung
ax2.plot(q, 10-0.8*q, 'b--', lw=1.8, label='D₁')
ax2.plot(q, 7-0.8*q, 'b-', lw=2.5, label='D₂')
ax2.plot(q, 1+0.7*q, 'r-', lw=2, label='S')
ax2.annotate('', xy=(3, 5), xytext=(5, 5), arrowprops=dict(arrowstyle='->', color='orange', lw=2.5))
ax2.set_title('Điện sạc đắt → Cầu EV GIẢM\n[Hàng bổ sung]', fontsize=11, fontweight='bold')
ax2.set_xlabel('Q'); ax2.set_ylabel('P')
ax2.legend(fontsize=9)
ax2.set_xlim(0, 10); ax2.set_ylim(0, 11)
ax2.grid(True, alpha=.25)

plt.suptitle('Kết luận Câu 2: Hàng bổ sung đắt → Lợi ích cận biên giảm', fontsize=12, fontweight='bold', y=0.02)
plt.tight_layout(rect=[0, 0.05, 1, 1])
plt.savefig(os.path.join(IMG_DIR, '02_substitute_complement.png'), dpi=150, bbox_inches='tight')
plt.show()
```

## 3. MR = MC – Quyết định sản lượng EcoDrive (Câu 3)

Lithium tăng → MC dịch lên → Điểm tối ưu MR=MC dịch sang trái → Giảm sản lượng từ Q₁ xuống Q₂.

```python
fig, ax = plt.subplots(figsize=(7.5, 5))
q = np.linspace(0.5, 10, 100)

ax.plot(q, 2+0.5*q, 'r--', lw=1.8, label='MC₁ (trước)')
ax.plot(q, 3.5+0.5*q, 'r-', lw=2.5, label='MC₂ (Lithium ↑)')
ax.plot(q, 9-0.8*q, 'g-', lw=2.5, label='MR')

q1 = (9-2)/1.3; q2 = (9-3.5)/1.3
ax.plot(q1, 2+0.5*q1, 'ko', ms=9)
ax.plot(q2, 3.5+0.5*q2, 'rs', ms=11)

ax.annotate('Q₁', (q1, 2+0.5*q1), xytext=(10, -20), textcoords='offset points', fontsize=12, fontweight='bold')
ax.annotate('Q₂', (q2, 3.5+0.5*q2), xytext=(10, 10), textcoords='offset points', fontsize=12, color='red', fontweight='bold')

ax.annotate('', xy=(q2, 1.2), xytext=(q1, 1.2), arrowprops=dict(arrowstyle='<->', color='purple', lw=2))
ax.text((q1+q2)/2, 0.6, 'Giảm sản lượng', ha='center', fontsize=11, color='purple', fontweight='bold')

ax.axvline(q1, color='gray', ls=':', alpha=.4)
ax.axvline(q2, color='red', ls=':', alpha=.4)

ax.set_xlabel('Sản lượng (Q)', fontsize=12)
ax.set_ylabel('Chi phí / Doanh thu', fontsize=12)
ax.set_title('Kết luận Câu 3: MC↑ → Giảm Q, Tăng P', fontsize=12, fontweight='bold')
ax.legend(fontsize=10, loc='upper right')
ax.set_xlim(0, 10); ax.set_ylim(0, 10)
ax.set_xticks(range(0, 11, 2))
ax.set_yticks(range(0, 11, 2))
ax.grid(True, alpha=.25)
plt.tight_layout()
plt.savefig(os.path.join(IMG_DIR, '03_mc_decision.png'), dpi=150, bbox_inches='tight')
plt.show()
```

## 4. Khác biệt hóa → Cầu ít co giãn (Câu 4)

- Trái: Không khác biệt hóa → đường cầu thoải → tăng giá nhẹ mất nhiều khách
- Phải: Có khác biệt hóa (R&D) → đường cầu dốc → tăng giá vẫn giữ được khách

```python
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(11, 4.5))
q = np.linspace(0, 10, 100)

# Không khác biệt hóa
ax1.plot(q, 8-0.4*q, 'b-', lw=2.5)
ax1.set_title('Không khác biệt hóa\nCầu co giãn (thoải)', fontsize=11, fontweight='bold')
ax1.annotate('Tăng giá nhẹ\n→ Mất NHIỀU khách', xy=(5, 6), xytext=(6.5, 8),
            arrowprops=dict(arrowstyle='->', color='red', lw=2), fontsize=10, color='red', fontweight='bold')
ax1.set_xlabel('Q'); ax1.set_ylabel('P')
ax1.set_xlim(0, 10); ax1.set_ylim(0, 10)
ax1.grid(True, alpha=.25)

# Có khác biệt hóa
ax2.plot(q, 10-1.5*q, 'b-', lw=2.5)
ax2.set_title('Có khác biệt hóa (R&D)\nCầu ít co giãn (dốc)', fontsize=11, fontweight='bold')
ax2.annotate('Tăng giá\n→ Mất ÍT khách', xy=(3.5, 4.75), xytext=(5.5, 7),
            arrowprops=dict(arrowstyle='->', color='green', lw=2), fontsize=10, color='green', fontweight='bold')
ax2.set_xlabel('Q'); ax2.set_ylabel('P')
ax2.set_xlim(0, 10); ax2.set_ylim(0, 10)
ax2.grid(True, alpha=.25)

plt.suptitle('Kết luận Câu 4: Khác biệt hóa → Cầu ít co giãn → Quyền lực cao', fontsize=12, fontweight='bold', y=0.02)
plt.tight_layout(rect=[0, 0.05, 1, 1])
plt.savefig(os.path.join(IMG_DIR, '04_differentiation.png'), dpi=150, bbox_inches='tight')
plt.show()
```

## 5. Chuyển đổi cấu trúc thị trường (Câu 4)

EcoDrive từ gần độc quyền → 10 đối thủ mới vào → Độc quyền nhóm → R&D tạo Cạnh tranh độc quyền.

```python
fig, ax = plt.subplots(figsize=(10, 3.5))
ax.set_xlim(0, 12); ax.set_ylim(0, 4); ax.axis('off')

boxes = [
    (0.3, 1.2, 3, 1.8, '#ffcccc', 'red', 'ĐỘC QUYỀN\nEcoDrive 60%\nMột hãng thống trị'),
    (4.5, 1.2, 3, 1.8, '#ffffcc', 'orange', 'ĐỘC QUYỀN NHÓM\nNhiều hãng lớn\nPhụ thuộc lẫn nhau'),
    (8.7, 1.2, 3, 1.8, '#ccffcc', 'green', 'CẠNH TRANH\nĐỘC QUYỀN\nSP khác biệt hóa'),
]

for x, y, w, h, fc, ec, txt in boxes:
    rect = mpatches.FancyBboxPatch((x, y), w, h, boxstyle="round,pad=0.1", facecolor=fc, edgecolor=ec, lw=2)
    ax.add_patch(rect)
    ax.text(x+w/2, y+h/2, txt, ha='center', va='center', fontsize=10, fontweight='bold')

ax.annotate('', xy=(4.3, 2.1), xytext=(3.5, 2.1), arrowprops=dict(arrowstyle='->', color='black', lw=2.5))
ax.annotate('', xy=(8.5, 2.1), xytext=(7.7, 2.1), arrowprops=dict(arrowstyle='->', color='black', lw=2.5))
ax.text(3.9, 3.2, '+10 đối thủ', ha='center', fontsize=9)
ax.text(8.1, 3.2, 'R&D + QC', ha='center', fontsize=9)

ax.set_title('Chuyển đổi cấu trúc thị trường xe điện Alpha', fontsize=12, fontweight='bold', pad=5)
plt.tight_layout()
plt.savefig(os.path.join(IMG_DIR, '05_market_structure.png'), dpi=150, bbox_inches='tight')
plt.show()
```
