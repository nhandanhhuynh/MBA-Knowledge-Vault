# Biểu đồ Tiểu luận BCG — DCMvn

Notebook tạo 4 hình ảnh chuyên nghiệp, in đen trắng vẫn rõ ràng (B&W-friendly).

- Hình 1: Sơ đồ tổ chức DCMvn

- Hình 4: Ma trận BCG — DCMvn (5 SBU)

```python
import matplotlib
import matplotlib.pyplot as plt
from matplotlib.patches import FancyBboxPatch, FancyArrowPatch
import matplotlib.patheffects as pe
import numpy as np
import os

OUTPUT_DIR = r"d:\01_Doc\00_Education\028_MBA\01_Courses\01_Active\00_BSKT\01_Quan Tri Hoc\07_AI-Workspace\Outputs"

plt.rcParams.update({
    'font.family': 'Arial',
    'font.sans-serif': ['Arial', 'DejaVu Sans'],
    'axes.unicode_minus': False,
    'figure.dpi': 150,
    'savefig.dpi': 300,
    'savefig.bbox': 'tight',
    'savefig.facecolor': 'white',
})

# B&W-friendly palette: dark enough to read when desaturated
C_DARK = '#1B2A3D'
C_MID  = '#3D5A80'
C_STAR = '#2A6496'   # prints as dark gray
C_COW  = '#1A7A4C'   # prints as medium-dark gray
C_QUES = '#B8860B'   # prints as medium gray
C_DOG  = '#8B2E2E'   # prints as dark gray
C_LIGHT = '#F0F2F5'
C_WHITE = '#FFFFFF'

print('Setup OK')
```

## Hình 1: Sơ đồ tổ chức DCMvn

```python
def create_org_chart(save_path=None):
    fig, ax = plt.subplots(figsize=(12, 7))
    ax.set_xlim(-1, 15)
    ax.set_ylim(0, 10)
    ax.axis('off')

    def box(x, y, w, h, text, fc, tc='white', fs=10, lw=2):
        rect = FancyBboxPatch((x - w/2, y - h/2), w, h,
                              boxstyle='round,pad=0.15',
                              facecolor=fc, edgecolor=C_DARK, linewidth=lw)
        ax.add_patch(rect)
        ax.text(x, y, text, ha='center', va='center', fontsize=fs,
                fontweight='bold', color=tc, linespacing=1.3)

    def arrow(x1, y1, x2, y2):
        ax.annotate('', xy=(x2, y2), xytext=(x1, y1),
                    arrowprops=dict(arrowstyle='->', color=C_DARK,
                                   lw=1.8, connectionstyle='arc3,rad=0'))

    # Level 0: CEO
    box(7, 9.0, 4.5, 1.0, 'TỔNG GIÁM ĐỐC\n(CEO)', C_DARK, 'white', 11)

    # Level 1: COO
    box(7, 7.0, 4.5, 1.0, 'GIÁM ĐỐC ĐIỀU HÀNH\n(COO)', C_MID, 'white', 11)
    arrow(7, 8.5, 7, 7.5)

    # Connector
    ax.plot([7, 7], [6.5, 5.8], color=C_DARK, lw=1.8)
    ax.plot([2, 12], [5.8, 5.8], color=C_DARK, lw=1.8)

    # Level 2: 4 departments — aligned horizontally
    depts = [
        (2.0,  'BỘ PHẬN MEP\n5 Trưởng nhóm\n~45 nhân viên', C_STAR),
        (5.5,  'BỘ PHẬN\nKIẾN TRÚC', '#2E7D32'),
        (8.8,  'BỘ PHẬN\nKỸ THUẬT SỐ\n1 Trưởng nhóm\n~5 nhân viên', C_QUES),
        (12.0, 'BỘ PHẬN\nHỖ TRỢ\n(NS, CNTT, HC)', '#555555'),
    ]
    for dx, txt, fc in depts:
        ax.plot([dx, dx], [5.8, 4.9], color=C_DARK, lw=1.8)
        box(dx, 3.8, 3.0, 2.0, txt, fc, 'white', 9)

    # Level 3: Staff under MEP and Digital
    box(2.0, 1.3, 3.0, 1.0, 'Kỹ sư mô hình MEP\nCấp 1 → Cấp 3', '#5B8DB8', 'white', 9)
    box(8.8, 1.3, 3.0, 1.0, 'Điều phối viên\nKỹ thuật số\nCấp 1 → Cấp 3', '#C9952A', 'white', 9)
    arrow(2.0, 2.8, 2.0, 1.8)
    arrow(8.8, 2.8, 8.8, 1.8)

    ax.set_title('SƠ ĐỒ TỔ CHỨC — CÔNG TY TNHH DCMvn',
                 fontsize=14, fontweight='bold', pad=15, color=C_DARK)

    plt.tight_layout()
    if save_path:
        plt.savefig(save_path, dpi=300, bbox_inches='tight', facecolor='white')
        print(f'Saved: {save_path}')
    plt.show()
    plt.close()
    return save_path

org_path = create_org_chart(os.path.join(OUTPUT_DIR, '_org_chart.png'))
```

## Hình 2: Ma trận BCG — Lý thuyết

```python
def create_bcg_theory(save_path=None):
    fig, ax = plt.subplots(figsize=(8, 7))
    ax.set_xlim(0, 10)
    ax.set_ylim(0, 10)

    # Quadrant fills — light, B&W-distinguishable via hatching
    quads = [
        ((5.05, 5.05), C_STAR, '///'),   # Star: top-right
        ((0.05, 5.05), C_QUES, '\\\\'),  # Question: top-left
        ((5.05, 0.05), C_COW,  '...'),    # Cow: bottom-right
        ((0.05, 0.05), C_DOG,  'xx'),     # Dog: bottom-left
    ]
    for (x, y), fc, hatch in quads:
        rect = FancyBboxPatch((x, y), 4.9, 4.9, boxstyle='round,pad=0.05',
                              facecolor=fc, alpha=0.12, edgecolor='none')
        ax.add_patch(rect)

    # Dividers
    ax.axhline(y=5, color=C_DARK, linewidth=2.5)
    ax.axvline(x=5, color=C_DARK, linewidth=2.5)

    # Labels with outline for readability
    outline = [pe.withStroke(linewidth=3, foreground='white')]
    labels = [
        (7.5, 8.0, 'NGÔI SAO\n(Star)', C_STAR, 22),
        (7.5, 6.0, 'Thị phần: CAO\nTăng trưởng: CAO', '#333', 10),
        (2.5, 8.0, 'DẤU HỎI\n(Question Mark)', C_QUES, 18),
        (2.5, 6.0, 'Thị phần: THẤP\nTăng trưởng: CAO', '#333', 10),
        (7.5, 3.5, 'BÒ SỮA\n(Cash Cow)', C_COW, 22),
        (7.5, 1.5, 'Thị phần: CAO\nTăng trưởng: THẤP', '#333', 10),
        (2.5, 3.5, 'CON CHÓ\n(Dog)', C_DOG, 22),
        (2.5, 1.5, 'Thị phần: THẤP\nTăng trưởng: THẤP', '#333', 10),
    ]
    for x, y, txt, c, fs in labels:
        ax.text(x, y, txt, ha='center', va='center', fontsize=fs,
                fontweight='bold', color=c, path_effects=outline)

    ax.set_xlabel('THỊ PHẦN TƯƠNG ĐỐI  →  Cao', fontsize=12, fontweight='bold',
                  labelpad=12, color=C_DARK)
    ax.set_ylabel('TỐC ĐỘ TĂNG TRƯỞNG THỊ TRƯỜNG  →  Cao', fontsize=12,
                  fontweight='bold', labelpad=12, color=C_DARK)
    ax.set_xticks([2.5, 7.5])
    ax.set_xticklabels(['Thấp', 'Cao'], fontsize=11, fontweight='bold')
    ax.set_yticks([2.5, 7.5])
    ax.set_yticklabels(['Thấp', 'Cao'], fontsize=11, fontweight='bold')
    ax.tick_params(length=0)
    ax.set_title('MA TRẬN BCG (Boston Consulting Group)', fontsize=15,
                 fontweight='bold', pad=18, color=C_DARK)
    for spine in ax.spines.values():
        spine.set_color(C_DARK)
        spine.set_linewidth(1.5)

    plt.tight_layout()
    if save_path:
        plt.savefig(save_path, dpi=300, bbox_inches='tight', facecolor='white')
        print(f'Saved: {save_path}')
    plt.show()
    plt.close()
    return save_path

bcg_theory_path = create_bcg_theory(os.path.join(OUTPUT_DIR, '_bcg_theory.png'))
```

## Hình 3: Chu kỳ sống sản phẩm

```python
def create_lifecycle(save_path=None):
    fig, ax = plt.subplots(figsize=(9, 5.5))

    x = np.linspace(0, 10, 300)
    y = 4 * np.exp(-0.5 * ((x - 6) / 2.2) ** 2)

    ax.plot(x, y, '-', linewidth=3.5, color=C_DARK, zorder=5)

    phases = [
        (0, 2.5, 'Giới thiệu\n(Introduction)', C_QUES, '////'),
        (2.5, 4.5, 'Tăng trưởng\n(Growth)', C_STAR, '\\\\\\\\'),
        (4.5, 7.5, 'Trưởng thành\n(Maturity)', C_COW, '....'),
        (7.5, 10, 'Suy thoái\n(Decline)', C_DOG, 'xxxx'),
    ]
    for x0, x1, label, color, hatch in phases:
        mask = (x >= x0) & (x <= x1)
        ax.fill_between(x[mask], y[mask], alpha=0.15, color=color)
        ax.axvline(x=x0, color=color, linewidth=1.2, linestyle='--', alpha=0.6)
        ax.text((x0 + x1) / 2, -0.7, label, ha='center', fontsize=10,
                fontweight='bold', color=color)

    sbu_labels = [
        (1.5, 0.8, 'SBU 4', C_QUES),
        (3.5, 2.9, 'SBU 2, 3', C_STAR),
        (6.0, 4.1, 'SBU 1', C_COW),
        (8.5, 1.8, 'SBU 5', C_DOG),
    ]
    for sx, sy, slbl, sc in sbu_labels:
        ax.annotate(slbl, xy=(sx, sy), fontsize=10, fontweight='bold', color=sc,
                    ha='center', bbox=dict(boxstyle='round,pad=0.3',
                    facecolor='white', edgecolor=sc, linewidth=1.5))

    ax.set_xlabel('Thời gian', fontsize=12, fontweight='bold', color=C_DARK)
    ax.set_ylabel('Doanh thu / Lợi nhuận', fontsize=12, fontweight='bold', color=C_DARK)
    ax.set_title('CHU KỲ SỐNG SẢN PHẨM VÀ VỊ TRÍ CÁC SBU CỦA DCMvn',
                 fontsize=13, fontweight='bold', pad=15, color=C_DARK)
    ax.set_xticks([])
    ax.set_yticks([])
    ax.set_ylim(-1.3, 5)
    ax.spines['top'].set_visible(False)
    ax.spines['right'].set_visible(False)
    ax.spines['left'].set_color(C_DARK)
    ax.spines['bottom'].set_color(C_DARK)

    plt.tight_layout()
    if save_path:
        plt.savefig(save_path, dpi=300, bbox_inches='tight', facecolor='white')
        print(f'Saved: {save_path}')
    plt.show()
    plt.close()
    return save_path

lifecycle_path = create_lifecycle(os.path.join(OUTPUT_DIR, '_lifecycle.png'))
```

## Hình 4: Ma trận BCG — DCMvn (5 SBU)

```python
def create_bcg_dcmvn(save_path=None):
    fig, ax = plt.subplots(figsize=(9, 8))
    ax.set_xlim(0, 10)
    ax.set_ylim(0, 10)

    # Quadrant backgrounds
    for (x, y), fc in [((5.05,5.05), C_STAR), ((0.05,5.05), C_QUES),
                        ((5.05,0.05), C_COW), ((0.05,0.05), C_DOG)]:
        ax.add_patch(FancyBboxPatch((x, y), 4.9, 4.9, boxstyle='round,pad=0.05',
                                    facecolor=fc, alpha=0.08, edgecolor='none'))

    ax.axhline(y=5, color=C_DARK, linewidth=2.5)
    ax.axvline(x=5, color=C_DARK, linewidth=2.5)

    # Quadrant labels (faint)
    outline = [pe.withStroke(linewidth=3, foreground='white')]
    for x, y, txt, c in [(7.5,9.5,'Ngôi sao (Star)',C_STAR),
                          (2.5,9.5,'Dấu hỏi (Question Mark)',C_QUES),
                          (7.5,0.5,'Bò sữa (Cash Cow)',C_COW),
                          (2.5,0.5,'Con chó (Dog)',C_DOG)]:
        ax.text(x, y, txt, ha='center', fontsize=11, fontweight='bold',
                color=c, alpha=0.5, path_effects=outline)

    # SBU bubbles — distinct markers for B&W
    sbus = [
        {'name': 'SBU 1\nMô hình MEP 3D', 'x': 7.5, 'y': 3.2,
         'size': 1100, 'color': C_COW, 'marker': 'o', 'edge': C_DARK},
        {'name': 'SBU 2\nBIM Kiến trúc', 'x': 6.2, 'y': 7.0,
         'size': 700, 'color': C_STAR, 'marker': 's', 'edge': C_DARK},
        {'name': 'SBU 3\nPhối hợp đa ngành', 'x': 7.0, 'y': 6.0,
         'size': 650, 'color': C_STAR, 'marker': 'D', 'edge': C_DARK},
        {'name': 'SBU 4\nScan to BIM & Số hóa', 'x': 2.8, 'y': 7.8,
         'size': 500, 'color': C_QUES, 'marker': '^', 'edge': C_DARK},
        {'name': 'SBU 5\nChuyển đổi 2D', 'x': 3.0, 'y': 2.5,
         'size': 400, 'color': C_DOG, 'marker': 'v', 'edge': C_DARK},
    ]

    for s in sbus:
        ax.scatter(s['x'], s['y'], s=s['size'], c=s['color'], alpha=0.75,
                   edgecolors=s['edge'], linewidth=2, zorder=5, marker=s['marker'])
        offset_y = -40 if s['y'] > 5 else 40
        ax.annotate(s['name'], (s['x'], s['y']),
                    textcoords='offset points', xytext=(0, offset_y),
                    ha='center', fontsize=9, fontweight='bold', color=s['color'],
                    bbox=dict(boxstyle='round,pad=0.3', facecolor='white',
                              edgecolor=s['edge'], linewidth=1.5, alpha=0.95))

    ax.set_xlabel('THỊ PHẦN TƯƠNG ĐỐI  →  Cao', fontsize=12, fontweight='bold',
                  labelpad=12, color=C_DARK)
    ax.set_ylabel('TỐC ĐỘ TĂNG TRƯỞNG  →  Cao', fontsize=12, fontweight='bold',
                  labelpad=12, color=C_DARK)
    ax.set_xticks([2.5, 7.5])
    ax.set_xticklabels(['Thấp', 'Cao'], fontsize=11, fontweight='bold')
    ax.set_yticks([2.5, 7.5])
    ax.set_yticklabels(['Thấp', 'Cao'], fontsize=11, fontweight='bold')
    ax.tick_params(length=0)
    ax.set_title('MA TRẬN BCG — DCMvn CO., Ltd', fontsize=15,
                 fontweight='bold', pad=18, color=C_DARK)
    for spine in ax.spines.values():
        spine.set_color(C_DARK)
        spine.set_linewidth(1.5)

    # Legend with distinct markers
    legend_items = [
        plt.Line2D([0],[0], marker='o', color='w', markerfacecolor=C_COW,
                   markeredgecolor=C_DARK, markersize=14, label='Bò sữa (Cash Cow)'),
        plt.Line2D([0],[0], marker='s', color='w', markerfacecolor=C_STAR,
                   markeredgecolor=C_DARK, markersize=12, label='Ngôi sao (Star) — Kiến trúc'),
        plt.Line2D([0],[0], marker='D', color='w', markerfacecolor=C_STAR,
                   markeredgecolor=C_DARK, markersize=11, label='Ngôi sao (Star) — Phối hợp'),
        plt.Line2D([0],[0], marker='^', color='w', markerfacecolor=C_QUES,
                   markeredgecolor=C_DARK, markersize=12, label='Dấu hỏi (Question Mark)'),
        plt.Line2D([0],[0], marker='v', color='w', markerfacecolor=C_DOG,
                   markeredgecolor=C_DARK, markersize=11, label='Con chó (Dog)'),
    ]
    ax.legend(handles=legend_items, loc='lower left', fontsize=9,
              framealpha=0.95, edgecolor=C_DARK)

    plt.tight_layout()
    if save_path:
        plt.savefig(save_path, dpi=300, bbox_inches='tight', facecolor='white')
        print(f'Saved: {save_path}')
    plt.show()
    plt.close()
    return save_path

bcg_dcmvn_path = create_bcg_dcmvn(os.path.join(OUTPUT_DIR, '_bcg_dcmvn.png'))
```
